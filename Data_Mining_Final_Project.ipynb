```python
#1 Load the libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#2 Load the dataset

df_supplychain = pd.read_csv("DataCoSupplyChainDataset.csv", encoding_errors = "ignore")
df_supplychain.head()
```

```python
#3 inspect dataset

df_supplychain.info()
```

```python
df_supplychain.describe(include="all")
```

```python
#4 Checking for missing values

print("Missing values per column:")
print(df_supplychain.isnull().sum())
```

```python
#5 Handeling of missing values

#Droping columns with too many missing values
df_supplychain = df_supplychain.drop(columns=["Order Zipcode", "Product Description"])

#filling missing values of other columns
df_supplychain["Customer Lname"] = df_supplychain["Customer Lname"].fillna("Unknown")
df_supplychain["Customer Zipcode"] =  df_supplychain["Customer Zipcode"].fillna(0)
```

```python
#after cleaning

print("Missing values after cleaning")
print(df_supplychain.isnull().sum())
```

```python
df_supplychain.describe(include="all")
```

```python
#6 Converting date columns to datetime

df_supplychain = df_supplychain.rename(columns={
    "order date (DateOrders)": "order_date",
    "shipping date (DateOrders)": "shipping_date"
})

df_supplychain["order_date"] = pd.to_datetime(df_supplychain["order_date"], errors="coerce")
df_supplychain["shipping_date"] = pd.to_datetime(df_supplychain["shipping_date"], errors="coerce")

df_supplychain[["order_date", "shipping_date"]].info()
```

```python
#7 Actual shipping duration

df_supplychain["delivery_days"] = (df_supplychain["shipping_date"] - df_supplychain["order_date"]).dt.days

# Delay vs schedule

df_supplychain["delay_vs_schedule"] = (df_supplychain["Days for shipping (real)"] - df_supplychain["Days for shipment (scheduled)"])
```

```python
#8 Extracting time based features

df_supplychain['order_year'] = df_supplychain['order_date'].dt.year
df_supplychain['order_month'] = df_supplychain['order_date'].dt.month
df_supplychain['order_day'] = df_supplychain['order_date'].dt.day
df_supplychain['order_hour'] = df_supplychain['order_date'].dt.hour
df_supplychain['order_minute'] = df_supplychain['order_date'].dt.minute

df_supplychain['shipping_year'] = df_supplychain['shipping_date'].dt.year
df_supplychain['shipping_month'] = df_supplychain['shipping_date'].dt.month
df_supplychain['shipping_day'] = df_supplychain['shipping_date'].dt.day
df_supplychain['shipping_hour'] = df_supplychain['shipping_date'].dt.hour
df_supplychain['shipping_minute'] = df_supplychain['shipping_date'].dt.minute
```

```python
df_supplychain[["delivery_days", "delay_vs_schedule", "order_year", "order_month", "shipping_year", "shipping_month"]].head()
```

```python
#9 Target distribution
print(df_supplychain["Late_delivery_risk"].value_counts())
print(df_supplychain["Late_delivery_risk"].value_counts(normalize=True))
```

```python
#10 EDA
# Visualation of dataset

# Late delivery vs Shipping mode

plt.figure(figsize=(8,5))
sns.countplot(data=df_supplychain, x="Shipping Mode", hue="Late_delivery_risk")
plt.title("Late Delivery Risk by Shipping Mode")
plt.xlabel("Shipping Mode")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
# Late delivery by region

late_by_region = (df_supplychain.groupby("Order Region")["Late_delivery_risk"].mean().sort_values())

plt.figure(figsize=(10,5))
late_by_region.plot(kind="bar")
plt.title("Average Late Delivery Risk by Region")
plt.ylabel("Late Delivery Rate")
plt.show()
```

```python
# Distribution of Delivery Time

sns.histplot(df_supplychain["delivery_days"], kde=True, bins=40)
plt.title("Distribution of Actual Delivery Days")
plt.xlabel("Delivery Days")
plt.show()
```

```python
# Delivery Time vs Lateness

plt.figure(figsize=(8,5))
sns.boxplot(data=df_supplychain, x="Late_delivery_risk", y="delivery_days")
plt.title("Delivery Days by Late Delivery Risk")
plt.xlabel("Late Delivery Risk (0 = On-time, 1 = Late)")
plt.ylabel("Delivery Days")
plt.show()
```

```python
# Delay vs Schedule Across Late/On-time Classes

sns.histplot(data=df_supplychain,x="delay_vs_schedule",hue="Late_delivery_risk",kde=True,bins=40,element="step")
plt.title("Delay vs Scheduled Days (Late vs On-time)")
plt.xlabel("Actual Days - Scheduled Days")
plt.show()
```

```python
#  Lateness by Customer Segment

plt.figure(figsize=(10,5))
sns.countplot(data=df_supplychain, x="Customer Segment", hue="Late_delivery_risk")
plt.title("Late Delivery Risk by Customer Segment")
plt.xlabel("Customer Segment")
plt.ylabel
plt.show()
```

```python
# Lateness vs delay_vs_schedule

plt.figure(figsize=(8,5))
sns.boxplot(data=df_supplychain, x="Late_delivery_risk", y="delay_vs_schedule")
plt.title("Delay vs Scheduled Days (Late vs On-time)")
plt.xlabel
plt.ylabel("Actual Days - Scheduled Days")
plt.show()
```
