# Hypothesis Testing

````python
import warnings
warnings.filterwarnings('ignore')
from scipy.special import chdtrc
from scipy.stats import t as t_dist

# Significance level
alpha = 0.05

print("="*80)
print("HYPOTHESIS TESTING: Relationships with Late Delivery Risk")
print("="*80)
````

## 1. Chi-Square Test: Categorical Variables vs Late Delivery Risk

````python
print("\n1. CHI-SQUARE TEST (Categorical Variables)")
print("-" * 80)
print("H0: Categorical variable is independent of Late Delivery Risk")
print("H1: Categorical variable is associated with Late Delivery Risk\n")

categorical_vars = ["Shipping Mode", "Order Region", "Customer Segment", "Product importance"]

chi_square_results = []

def chi_square_test(contingency_table):
    """Calculate chi-square statistic using numpy and pandas"""
    obs = contingency_table.values.astype(float)
    row_totals = obs.sum(axis=1, keepdims=True)
    col_totals = obs.sum(axis=0, keepdims=True)
    grand_total = obs.sum()
    
    expected = (row_totals * col_totals) / grand_total
    chi2 = np.sum((obs - expected) ** 2 / expected)
    dof = (obs.shape[0] - 1) * (obs.shape[1] - 1)
    p_value = chdtrc(dof, chi2)
    
    return chi2, p_value, dof

for var in categorical_vars:
    if var in df_supplychain.columns:
        contingency_table = pd.crosstab(df_supplychain[var], df_supplychain["Late_delivery_risk"])
        chi2, p_value, dof = chi_square_test(contingency_table)
        
        result = {
            'Variable': var,
            'Chi-Square': f"{chi2:.4f}",
            'p-value': f"{p_value:.6f}",
            'Significant': 'Yes' if p_value < alpha else 'No'
        }
        chi_square_results.append(result)
        
        print(f"Variable: {var}")
        print(f"  Chi-Square Statistic: {chi2:.4f}")
        print(f"  P-value: {p_value:.6f}")
        print(f"  Degrees of Freedom: {dof}")
        print(f"  Result: {'REJECT H0 - Significant' if p_value < alpha else 'FAIL TO REJECT H0 - Not Significant'} (Î± = {alpha})")
        print()

chi_square_df = pd.DataFrame(chi_square_results)
print("\nSummary Table:")
print(chi_square_df.to_string(index=False))
````

## 2. Independent t-test: Continuous Variables vs Late Delivery Risk

````python
print("\n2. INDEPENDENT T-TEST (Continuous Variables)")
print("-" * 80)
print("H0: Mean of continuous variable is equal for Late=0 and Late=1")
print("H1: Mean of continuous variable differs between Late=0 and Late=1\n")

continuous_vars = ['Days for shipment (scheduled)', 'Days for shipping (real)', 
                   'Product quantity ordered', 'Product price', 'delivery_days', 'delay_vs_schedule']

ttest_results = []

def independent_ttest(group1, group2):
    """Calculate t-test using numpy"""
    n1, n2 = len(group1), len(group2)
    mean1, mean2 = np.mean(group1), np.mean(group2)
    var1, var2 = np.var(group1, ddof=1), np.var(group2, ddof=1)
    
    se = np.sqrt((var1/n1) + (var2/n2))
    t_stat = (mean1 - mean2) / se if se != 0 else 0
    df = n1 + n2 - 2
    
    # Two-tailed p-value using t-distribution
    p_value = 2 * (1 - t_dist.cdf(abs(t_stat), df))
    
    return t_stat, p_value

for var in continuous_vars:
    if var in df_supplychain.columns:
        on_time = df_supplychain[df_supplychain['Late_delivery_risk'] == 0][var].dropna().values
        late = df_supplychain[df_supplychain['Late_delivery_risk'] == 1][var].dropna().values
        
        t_stat, p_value = independent_ttest(on_time, late)
        
        result = {
            'Variable': var,
            't-statistic': f"{t_stat:.4f}",
            'p-value': f"{p_value:.6f}",
            'Mean (On-time)': f"{np.mean(on_time):.2f}",
            'Mean (Late)': f"{np.mean(late):.2f}",
            'Significant': 'Yes' if p_value < alpha else 'No'
        }
        ttest_results.append(result)
        
        print(f"Variable: {var}")
        print(f"  Mean (On-time, Late=0): {np.mean(on_time):.2f}")
        print(f"  Mean (Late, Late=1): {np.mean(late):.2f}")
        print(f"  t-statistic: {t_stat:.4f}")
        print(f"  P-value: {p_value:.6f}")
        print(f"  Result: {'REJECT H0 - Significant' if p_value < alpha else 'FAIL TO REJECT H0 - Not Significant'} (Î± = {alpha})")
        print()

ttest_df = pd.DataFrame(ttest_results)
print("\nSummary Table:")
print(ttest_df.to_string(index=False))
````

## 4. Point-Biserial Correlation (Strength of Association)

### Understanding Point-Biserial Correlation

**What is it?**
Point-Biserial Correlation measures the **strength and direction of the relationship** between:
- A **continuous variable** (e.g., delivery days, product price)
- A **binary variable** (e.g., Late delivery: Yes/No, 0/1)

It's essentially a **Pearson correlation** adapted for binary data.

**When to use it?**
- You want to measure **correlation strength** (not just if it's significant)
- You have one continuous and one binary variable
- Similar to correlation coefficient but appropriate for binary outcomes

**How does it work?**
1. Convert binary variable to 0 and 1
2. Calculate correlation coefficient between continuous and binary variables
3. Ranges from **-1 to +1** (like Pearson correlation)
4. Also provides p-value for statistical significance

**Formula (Conceptual):**
$$r_{pb} = \frac{M_1 - M_0}{S_n} \sqrt{\frac{n_1 n_0}{n^2}}$$

Where:
- $M_1$ = mean of continuous variable when binary = 1
- $M_0$ = mean of continuous variable when binary = 0
- $S_n$ = standard deviation of continuous variable
- $n_1, n_0$ = counts of binary = 1 and 0

**Interpretation:**
- **r = 0.7 to 1.0**: Strong positive correlation
- **r = 0.4 to 0.7**: Medium positive correlation
- **r = 0.1 to 0.4**: Weak positive correlation
- **r = 0 to 0.1**: Very weak or no correlation
- **Negative values** indicate inverse relationships

**Simple Example:**
```
Late Delivery (Binary):    [0, 0, 0, 1, 1, 1]
Delivery Days (Continuous): [2, 2.5, 3, 4, 4.5, 5]

Correlation = 0.95 (very strong positive)
Interpretation: Higher delivery days â†’ Higher chance of late delivery
```

**Why is it useful?**
- Identifies which continuous variables **best predict** your binary outcome
- Helps prioritize features for machine learning models
- Stronger correlation = Better predictor of late delivery

````python
print("\n3. POINT-BISERIAL CORRELATION (Strength of Association)")
print("-" * 80)
print("Measures correlation between continuous variable and binary target (Late_delivery_risk)\n")

correlation_results = []

def point_biserial_correlation(binary_var, continuous_var):
    """Calculate point-biserial correlation using numpy and pandas"""
    # Remove NaN values
    mask = ~(np.isnan(binary_var) | np.isnan(continuous_var))
    binary = binary_var[mask]
    continuous = continuous_var[mask]
    
    # Mean and std of continuous variable
    M0 = np.mean(continuous[binary == 0])
    M1 = np.mean(continuous[binary == 1])
    Sn = np.std(continuous, ddof=1)
    
    n0 = np.sum(binary == 0)
    n1 = np.sum(binary == 1)
    n = len(binary)
    
    # Point-biserial correlation
    if Sn == 0:
        corr = 0
    else:
        corr = ((M1 - M0) / Sn) * np.sqrt((n0 * n1) / (n ** 2))
    
    # For p-value, use t-test on correlation
    t_stat = corr * np.sqrt(n - 2) / np.sqrt(1 - corr**2) if abs(corr) < 1 else 0
    p_value = 2 * (1 - t_dist.cdf(abs(t_stat), n - 2)) if abs(corr) < 1 else 0
    
    return corr, p_value

for var in continuous_vars:
    if var in df_supplychain.columns:
        binary = df_supplychain['Late_delivery_risk'].values
        continuous = df_supplychain[var].values
        
        corr, p_value = point_biserial_correlation(binary, continuous)
        
        result = {
            'Variable': var,
            'Correlation': f"{corr:.4f}",
            'p-value': f"{p_value:.6f}",
            'Effect Size': 'Weak' if abs(corr) < 0.3 else ('Medium' if abs(corr) < 0.5 else 'Strong'),
            'Significant': 'Yes' if p_value < alpha else 'No'
        }
        correlation_results.append(result)
        
        print(f"Variable: {var}")
        print(f"  Point-Biserial Correlation: {corr:.4f}")
        print(f"  P-value: {p_value:.6f}")
        print(f"  Effect Size: {'Weak' if abs(corr) < 0.3 else ('Medium' if abs(corr) < 0.5 else 'Strong')}")
        print(f"  Significant: {'Yes' if p_value < alpha else 'No'}")
        print()

correlation_df = pd.DataFrame(correlation_results)
print("\nSummary Table:")
print(correlation_df.to_string(index=False))
````

## 5. ANOVA Test for Multiple Groups

````python
print("\n4. ANOVA TEST (Testing differences across multiple groups)")
print("-" * 80)
print("H0: Mean values are equal across all groups")
print("H1: At least one group mean differs significantly\n")

anova_results = []

def one_way_anova(*groups):
    """Calculate one-way ANOVA using numpy"""
    # Flatten all groups
    all_data = np.concatenate(groups)
    
    # Number of groups
    k = len(groups)
    
    # Sample sizes
    n_i = np.array([len(g) for g in groups])
    n = np.sum(n_i)
    
    # Grand mean
    grand_mean = np.mean(all_data)
    
    # Between-group sum of squares
    ss_between = np.sum(n_i * (np.array([np.mean(g) for g in groups]) - grand_mean) ** 2)
    
    # Within-group sum of squares
    ss_within = np.sum([np.sum((g - np.mean(g)) ** 2) for g in groups])
    
    # Degrees of freedom
    df_between = k - 1
    df_within = n - k
    
    # Mean squares
    ms_between = ss_between / df_between
    ms_within = ss_within / df_within if df_within > 0 else 1
    
    # F-statistic
    f_stat = ms_between / ms_within if ms_within > 0 else 0
    
    # P-value using F-distribution
    from scipy.stats import f as f_dist
    p_value = 1 - f_dist.cdf(f_stat, df_between, df_within)
    
    return f_stat, p_value

if 'Shipping Mode' in df_supplychain.columns:
    groups = [group['delivery_days'].dropna().values for name, group in 
              df_supplychain.groupby('Shipping Mode')]
    
    f_stat, p_value = one_way_anova(*groups)
    
    result = {
        'Test': 'Delivery Days by Shipping Mode',
        'F-statistic': f"{f_stat:.4f}",
        'p-value': f"{p_value:.6f}",
        'Significant': 'Yes' if p_value < alpha else 'No'
    }
    anova_results.append(result)
    
    print(f"ANOVA: Delivery Days across Shipping Modes")
    print(f"  F-statistic: {f_stat:.4f}")
    print(f"  P-value: {p_value:.6f}")
    print(f"  Result: {'REJECT H0 - Significant' if p_value < alpha else 'FAIL TO REJECT H0 - Not Significant'} (Î± = {alpha})")
    print()

if 'Customer Segment' in df_supplychain.columns:
    groups = [group['delivery_days'].dropna().values for name, group in 
              df_supplychain.groupby('Customer Segment')]
    
    f_stat, p_value = one_way_anova(*groups)
    
    result = {
        'Test': 'Delivery Days by Customer Segment',
        'F-statistic': f"{f_stat:.4f}",
        'p-value': f"{p_value:.6f}",
        'Significant': 'Yes' if p_value < alpha else 'No'
    }
    anova_results.append(result)
    
    print(f"ANOVA: Delivery Days across Customer Segments")
    print(f"  F-statistic: {f_stat:.4f}")
    print(f"  P-value: {p_value:.6f}")
    print(f"  Result: {'REJECT H0 - Significant' if p_value < alpha else 'FAIL TO REJECT H0 - Not Significant'} (Î± = {alpha})")
    print()

if anova_results:
    anova_df = pd.DataFrame(anova_results)
    print("\nANOVA Summary Table:")
    print(anova_df.to_string(index=False))
````
