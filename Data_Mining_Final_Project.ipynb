Hypothesis Testing

T Test: delivery_days (On-time vs Late)

```python
# Hypothesis tests: t-test 
import scipy.stats as stats

# Prepare numeric groups for delivery days split by lateness
on_time = df_supplychain.loc[df_supplychain["Late_delivery_risk"]==0, "delivery_days"].dropna()
late = df_supplychain.loc[df_supplychain["Late_delivery_risk"]==1, "delivery_days"].dropna()

print("=== T-test : delivery_days (On-time vs Late) ===")
print(f"Counts: on_time={len(on_time)}, late={len(late)}")

# Check normality (Shapiro) on a sample if the groups are large
try:
    sh_on = stats.shapiro(on_time.sample(500, random_state=0)) if len(on_time) > 500 else stats.shapiro(on_time)
    sh_late = stats.shapiro(late.sample(500, random_state=0)) if len(late) > 500 else stats.shapiro(late)
    print(f"Shapiro on_time: stat={sh_on.statistic:.4f}, p={sh_on.pvalue:.4g}")
    print(f"Shapiro late: stat={sh_late.statistic:.4f}, p={sh_late.pvalue:.4g}")
except Exception as e:
    print("Shapiro test failed or not applicable:", e)

# Test equality of variances (Levene)
lev = stats.levene(on_time, late)
print(f"Levene test: stat={lev.statistic:.4f}, p={lev.pvalue:.4g}")

# Choose parametric or non-parametric test based on normality
use_ttest = False
if ('sh_on' in locals() and 'sh_late' in locals()):
    use_ttest = (sh_on.pvalue > 0.05) and (sh_late.pvalue > 0.05)
if use_ttest:
    print("Both groups approx normal -> performing t-test")
    tstat, pval = stats.ttest_ind(on_time, late, equal_var=(lev.pvalue > 0.05))
    print(f"t-statistic={tstat:.4f}, p-value={pval:.4g}")
else:
    print("Non-normal or uncertain -> performing Mann-Whitney U test")
    ustat, upval = stats.mannwhitneyu(on_time, late, alternative='two-sided')
    print(f"U-statistic={ustat:.4f}, p-value={upval:.4g}")
```

Chi-Square test: Shipping Mode vs Late_delivery_risk

```python
print("\n=== Chi-square Test: Shipping Mode vs Late_delivery_risk ===")
# Chi-square test: Shipping Mode vs Late_delivery_risk
cont = pd.crosstab(df_supplychain["Shipping Mode"], df_supplychain["Late_delivery_risk"])
print('\nContingency table:')
print(cont)
chi2, p, dof, expected = stats.chi2_contingency(cont)
print(f"Chi-square: chi2={chi2:.4f}, p-value={p:.4g}, dof={dof}")
```

ANOVA TEST: delivery_days by Shipping Mode

```python
print("\n=== ANOVA / Kruskal-Wallis: delivery_days by Shipping Mode ===")
# ANOVA: compare delivery_days across Shipping Mode groups (or non-parametric Kruskal-Wallis)
modes = df_supplychain["Shipping Mode"].dropna().unique()
groups = [df_supplychain.loc[df_supplychain["Shipping Mode"]==m, "delivery_days"].dropna() for m in modes]
valid_groups = [g for g in groups if len(g) > 0]
if len(valid_groups) >= 2:
    # Check approximate normality per group (sample if large)
    normal_flags = []
    for g in valid_groups:
        try:
            sh = stats.shapiro(g.sample(500, random_state=0)) if len(g) > 500 else stats.shapiro(g)
            normal_flags.append(sh.pvalue > 0.05)
        except Exception:
            normal_flags.append(False)
    if all(normal_flags):
        fstat, fp = stats.f_oneway(*valid_groups)
        print(f"\nANOVA f-statistic={fstat:.4f}, p-value={fp:.4g}")
    else:
        kw, kp = stats.kruskal(*valid_groups)
        print(f"\nKruskal-Wallis H-statistic={kw:.4f}, p-value={kp:.4g}")
else:
    print("Not enough groups with data for ANOVA / Kruskal-Wallis")
```

**PCA Test**

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Select numeric features
numeric_df = df_supplychain.select_dtypes(include=[np.number]).copy()

# Drop target & ID columns
cols_to_drop = [
    "Late_delivery_risk",  
    "Customer Id",
    "Order Id",
    "Order Item Id",
    "Order Customer Id",
    "Order Item Cardprod Id",
    "Product Card Id",
    "Product Category Id",
    "Category Id",
    "Department Id",
    "Customer Zipcode"      
]

numeric_df = numeric_df.drop(columns=[c for c in cols_to_drop if c in numeric_df.columns])

print("Numeric features used for PCA:", numeric_df.columns.tolist())
```

```python
# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(numeric_df)

# Fit PCA
pca = PCA(n_components=None, random_state=42)
X_pca = pca.fit_transform(X_scaled)

# variance ratio
var_ratio = pca.explained_variance_ratio_
print("Explained variance by component:")
for i, var in enumerate(var_ratio, start=1):
    print(f"PC{i}: {var:.4f}, cumulative: {var_ratio[:i].sum():.4f}")
```

```python
plt.figure(figsize=(8,5))
plt.plot(np.arange(1, len(var_ratio)+1), var_ratio, marker="o")
plt.title("PCA Scree Plot")
plt.xlabel("Principal Component")
plt.ylabel("Explained Variance Ratio")
plt.xticks(np.arange(1, len(var_ratio)+1))
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# PCA components_ shape
loadings = pd.DataFrame(pca.components_.T, columns=[f"PC{i}" for i in range(1, len(var_ratio)+1)], index=numeric_df.columns)

# Show top variables (by absolute loading) for first 3 PCs
for pc in ["PC1", "PC2", "PC3"]:
    print(f"\nTop variables for {pc}:")
    print(loadings[pc].abs().sort_values(ascending=False).head(10))
```
